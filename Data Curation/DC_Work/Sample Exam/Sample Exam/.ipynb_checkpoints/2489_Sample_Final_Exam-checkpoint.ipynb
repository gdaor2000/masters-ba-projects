### Question 1 (10 points)

Read the `titanic.xlsx` and shows how many passnegers records in the data.

Due to the errors in the history archives, there are several problems you need to address first in order to obtain the correct data:

1. In the column of *`sibsp`*, the value of 1 is mistakenly recorded as -1
2. In the column of *`survived`*, the value of 0 is mistakenly recorded as NaN

```python
titanic_data = pd.read_excel('titanic.xlsx', header = 20).set_index('Unnamed: 0')
titanic_data
```

```python
titanic_data['sibsp'] = titanic_data['sibsp'].abs()
titanic_data['survived'].fillna(0)

titanic_data
```

### Question 2 (5 points)

Show how many male and female passengers in terms of percentage:

```python
maleDF = titanic_data[titanic_data['sex'] == 'male']

percentage_male = (len(maleDF)/ len(titanic_data)) * 100

femaleDF = titanic_data[titanic_data['sex'] == 'female']

percentage_female = (len(femaleDF)/ len(titanic_data)) * 100

print(f'Male passengers in terms of percentage: {round(percentage_male,2)}%')
print(f'Female passengers in terms of percentage: {round(percentage_female,2)}%')
```

### Question 3 (5 points)
Show the average number of siblings/spouses for passengers embarked from Southampton

```python
print(round(titanic_data.loc[titanic_data.embark_town == 'Southampton']['sibsp'].mean(),2))
```

### Question 4 (5 points)

Show the median age of passengers that are adult male:

```python
print(round(titanic_data.loc[titanic_data.adult_male == True]['age'].median(),2))
```

### Question 5 (5 points)

Show the mean difference of fares between First Class passengers and Third Class passengers: 

```python
first_class = titanic_data.loc[titanic_data ['class'] == 'First']['fare'].mean()
third_class = titanic_data.loc[titanic_data ['class'] == 'Third']['fare'].mean()
avg_dif = first_class - third_class

round(avg_dif,2)
```

### Question 6 (5 points)

Show the survival status of passengers with the top 10 highest fare:

```python
survival_status_top_10 = titanic_data.sort_values(by = 'fare', ascending=False).head(10)

top_10 = survival_status_top_10[['survived', 'fare']]

top_10
```

### Question 7 (5 points)
Show the survival rate of men, women and children, respectively:

```python
man_survived = titanic_data[(titanic_data['who']=='man') & (titanic_data['survived']== 1)]
man = titanic_data[titanic_data['who']=='man']

ratio_man = (len(man_survived)/len(man))*100

child_survived = titanic_data[(titanic_data['who']=='child') & (titanic_data['survived']== 1)]
child = titanic_data[titanic_data['who']=='child']

ratio_child = (len(child_survived)/len(child))*100

woman_survived = titanic_data[(titanic_data['who']=='woman') & (titanic_data['survived']== 1)]
woman = titanic_data[titanic_data['who']=='woman']

ratio_woman = (len(woman_survived)/len(woman))*100

print(f'Survival Rate of men: {round(ratio_man,2)}')
print(f'Survival Rate of woman: {round(ratio_woman,2)}')
print(f'Survival Rate of child: {round(ratio_child,2)}')
```

### Question 8 (5 points)

Did number of siblings increase the odds of survival of different sex (male, female)? Show the survival rate given different number of siblings for male and female passengers: 

```python
male_sib = man[['survived','sibsp']].groupby('sibsp').sum()

male_total = man[['survived','sibsp']].groupby('sibsp').count()

male_rate = (male_sib/male_total) * 100

female_sib = woman[['survived','sibsp']].groupby('sibsp').sum()

female_total = woman[['survived','sibsp']].groupby('sibsp').count()

female_rate = (female_sib/female_total) * 100

print(f'Male survival rate per sibling: \n\n{male_rate}')
print(f'\nFemale survival rate per sibling: \n\n{female_rate}')  
```

### Question 9 (5 points)

Show the number of passengers across groups of cabin calss, gender, port of embarkation:

### Question 10 (10 points)

Now you need to show the top 5 demographic groups with the highest survival rate. The demographic group should consider the following three characteristics: *sex, ticket class and whether the passenger is alone*:

```python
demographic = titanic_data[['sex','class','alone','survived']].groupby(['sex','class','alone']).sum()
demographic_total = titanic_data[['sex','class','alone','survived']].groupby(['sex','class','alone']).count()

demographic_rate = round((demographic/demographic_total)*100,2)

demographic_sorted = demographic_rate.sort_values(by='survived',ascending = False).head(5)
demographic_sorted
```

### Question 11 (10 points)

Now show the survival rate of passengers across 5 quintile levels of ticket fares, namely, what is the survival rate among passengers with the top 20%, 20-40%, 40-60%, 60-80% and bottom 20% ticket fares. Can you identify the difference between male and female passengers?

```python
titanic_data['quantile'] = pd.qcut(titanic_data['fare'],q=[0,0.2,0.4,0.6,0.8,1],labels=['0% - 20%', '20%-40%', '40%-60%', '60%-80%','80%-100%'])
titanic_data.groupby(['sex','quantile'])[['survived']].sum()/titanic_data.groupby(['sex','quantile'])[['survived']].count()
```
