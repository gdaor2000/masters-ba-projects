```python
#First importa all the libraries that we need
import pandas as pd
import numpy as np
%pylab inline
```

```python
#Read all the files that we need
gdp = pd.read_csv('gdp_csv.csv')
population = pd.read_csv('population.csv')
country = pd.read_csv('country_definitions.csv')
athlete = pd.read_csv('athlete_events.csv')
```

Now let's start cleaning the data

```python
#Let's see the missing data that we have
print(athlete.isnull().sum())
```

```python
#We are going to fill the NAN in the column Medals
athlete['Medal'].fillna('None', inplace=True)
```

```python
# Now we are going to see if the NOC is connected to only one team
print(athlete.loc[:, ['NOC', 'Team']].drop_duplicates()['NOC'].value_counts().head())
```

```python
#See the information that we have in country
country
```

### Now let's merge the files athletes and countries



```python
# merging the athelete file with the country
olympics= athlete.merge(country,
                                left_on = 'NOC',
                                right_on = 'NOC',
                                how = 'left')
olympics
```

```python
# See if we have NOC's that dont match in the column country
olympics.loc[olympics['region'].isnull(),['NOC', 'Team']].drop_duplicates()
```

```python
# Replace missing Teams by the values above.
olympics['region'] = np.where(olympics['NOC']=='SGP', 'Singapore', olympics['region'])
olympics['region'] = np.where(olympics['NOC']=='ROT', 'Refugee Olympic Athletes', olympics['region'])
olympics['region'] = np.where(olympics['NOC']=='UNK', 'Unknown', olympics['region'])
olympics['region'] = np.where(olympics['NOC']=='TUV', 'Tuvalu', olympics['region'])


# Put these values from Country into Team
olympics.drop('Team', axis = 1, inplace = True)
olympics.rename(columns = {'region': 'Team'}, inplace = True)
olympics
```

```python
#Dropping the column notes
olympics= olympics.drop('notes', axis=1)
olympics
```

### Now for the gdp file

```python
#Show the file gdp
gdp
```

```python
#Change the name of the columns Values and year
gdp.rename(columns={'Year':'Year','Value':'GDP'}, inplace=True)
gdp
```

```python
# convert the year column to numeric
gdp['Year'] = pd.to_numeric(gdp['Year'], errors='coerce')
gdp
```

Now we are going to see what is the best column to use for the joining of the two tables, the NOC and country code or NOC and country Name

```python
#Checking if NOCs in the olympics data match with those in the Country Code
len(list(set(olympics['NOC'].unique()) - set(gdp['Country Code'].unique())))
```

So, 111 NOCs in the olympics dataset have a representation in the gdp data country codes.

```python
#Lets try to use the name of the country is a better way to merge?
len(list(set(olympics['Team'].unique()) - set(gdp['Country Name'].unique())))
```

Since we only have 33 matched, so maybe what we can do is to add a country code for each Team in the olympics dataset first and then merge using the Country Code.

```python
# Merge the file gd with the olympics
olympics_ccode = olympics.merge(gdp[['Country Name', 'Country Code']].drop_duplicates(),
                                            left_on = 'Team',
                                            right_on = 'Country Name',
                                            how = 'left')

olympics_ccode.drop('Country Name', axis = 1, inplace = True)
olympics_ccode
```

```python
#Use the apply function to only update the 'Country Code' column in these specific case
olympics_ccode['Country Code'] = olympics_ccode.apply(lambda row: 'USA' if row['NOC'] == 'USA' else row['Country Code'], axis=1)
olympics_ccode
```

```python
# Merge the gdp and the olympics_code
olympics_gdp = olympics_ccode.merge(gdp, left_on=['NOC', 'Year'], right_on=['Country Code', 'Year'], how='left')
olympics_gdp.drop(['Country Name','Country Code_y'], axis = 1, inplace = True)
olympics_gdp.rename(columns = {'Country Code_x': 'Country Code'}, inplace = True)
olympics_gdp
```

# Now lets do for the population set

```python
#See the file population
population
```

```python
#Drop all the rows that are all blanks
population = population.dropna(how='all')
population
```

```python
#Transform the column geo into upper letters
population['geo'] = population['geo'].str.upper()
population
```

```python
#Checking if NOCs in the olympics data match with those in the Country Code
len(list(set(olympics_gdp['NOC'].unique()) - set(population['geo'].unique())))
```

```python
#Lets try to use the name of the country is a better way to merge?
len(list(set(olympics_gdp['Team'].unique()) - set(population['name'].unique())))
```

```python
# Merge the olympics_gdp and the population
olympics_complete = olympics_gdp.merge(
    population,
    left_on=['Country Code', 'Year'],
    right_on=['geo', 'time'],
    how='left'
)
olympics_complete
```

```python
#Drop the columns geo and time
olympics_complete.drop(['geo','time'], axis = 1, inplace = True)
olympics_complete
```

## Now that we have all the data in one data set, we are going to arrange the information

```python
#Create a column that say 1 if they win a medal and 0 if they dont win a medal
olympics_complete['Medal_Won'] = np.where(olympics_complete.loc[:,'Medal'] == 'None', 0, 1)
olympics_complete
```

```python
#Identify team events and update 'Team Sports' column to 'yes'
identify_team_events = pd.pivot_table(olympics_complete,
                                       index=['Team', 'Year', 'Event'],
                                       columns='Medal',
                                       values='Medal_Won',
                                       aggfunc='sum',
                                       fill_value=0).drop('None', axis=1).reset_index()

identify_team_events = identify_team_events.loc[identify_team_events['Gold'] > 1, :]

team_sports = identify_team_events['Event'].unique()

#Update 'Team Sports' column in the original datset
olympics_complete.loc[olympics_complete['Event'].isin(team_sports), 'Team Sports'] = 'Yes'
olympics_complete
```

```python
#Remove the individual sports identified in the team_sports
remove_sports = ["Gymnastics Women's Balance Beam", "Gymnastics Men's Horizontal Bar",
                 "Swimming Women's 100 metres Freestyle", "Swimming Men's 50 metres Freestyle","Alpinism Mixed Alpinism","Art Competitions Mixed Architecture","Art Competitions Mixed Architecture, Architectural Designs","Art Competitions Mixed Architecture", "Designs For Town Planning","Motorboating Mixed C-Class","Speed Skating Men's 1,500 metres","Athletics MenAthletics Mens Pole Vaults Pole Vault","Cross Country Skiing Mens 10/10 kilometres Pursuit","Figure Skating Mixed Ice Dancing"]

team_sports = list(set(team_sports) - set(remove_sports))
```

```python
#Download the file
olympics_complete.to_csv("olympics_complete.csv", index=False)
```
